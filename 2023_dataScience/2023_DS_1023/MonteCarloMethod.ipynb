```python
# ライブラリの読み込み
import math
import random

# 初期化
n = 100000000
cnt_in = 0
cnt_out = 0

# シミュレーション
for i in range(n):
  # 0.0 - 1.0 のランダムな「雨粒」
  x = random.random()
  y = random.random()
  # 4分円の中に落ちた数をカウント
  r = math.sqrt(x**2 + y**2)
  if r <= 1.0:
      cnt_in += 1
  else:
      cnt_out += 1

# 円周率の近似値
Pi = cnt_in / n * 4

# 結果の表示
print("Num：", n)
print("IN：", cnt_in)
print("OUT：", cnt_out)
print("Pi：", Pi)
```
