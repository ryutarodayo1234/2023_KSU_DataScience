```python
# pandasをインポート
import pandas as pd

# サンプルデータ Tips の読み込み
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# データフレームの表示
df
```

```python
# データフレームの基本情報
print('行数・列数：', df.shape)
print('項目名：', df.columns)
print('各列のデータ型：\n', df.dtypes)
```

```python
# 要約統計量を表示
df.describe()
```

```python
# 相関係数を表示
df.corr(numeric_only=True)
```

```python
# データを tip の順で並べ替え（ 降順 ）
df_sort = df.sort_values( 'tip', ascending=False )
df_sort
```

```python
# Sunの行のみ抽出
df_female = df.query( ' day == "Sun" ' )
df_female
```

```python
# Sun に tip 5以上で来た客のみ抽出
df_male_smoker = df.query( ' day == "Sun" and tip >= 5 ' )
df_male_smoker
```

```python
# Sat に tip 5以上で来た客のみ抽出
df_male_smoker = df.query( ' day == "Sat" and tip >= 5 ' )
df_male_smoker
```

```python
# クロス集計 day別のtimeの数
cross_table = pd.crosstab( df['day'] , df['time'] )
cross_table
```

```python
# クロス集計 time別のsmokerの数
cross_table = pd.crosstab( df['time'] , df['smoker'] )
cross_table
```

```python
# クロス集計 day別のtimeの数 正規化
cross_table_n = pd.crosstab( df['day'] , df['time'] , normalize='index' )
cross_table_n
```

```python
df_day_counts = df['day'].value_counts()
df_day_counts
```

```python
df_day_mean = df.groupby('day').mean()
df_day_mean
```

```python
# size 2 に tip 5以上で来た客のみ抽出
df_male_smoker = df.query( ' size == 2 and tip >= 5 ' )
df_male_smoker
```

```python
# size 2 に tip 5以上でtime dinnerでday Sunで来たNo smoking客のみ抽出
df_male_smoker = df.query( ' size == 2 and tip >= 5 and time == "Dinner" and day == "Sun" and smoker == "No" ' )
df_male_smoker
```
